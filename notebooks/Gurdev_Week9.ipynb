Intro to ML using Linear Regressions



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tdf = pd.read_csv("data9_tips.csv")
tdf.head(3)
```

```python
import patsy as pts
```

```python
y,x = pts.dmatrices('tip ~ total_bill + size + C(time)', data = tdf, return_type = 'dataframe')
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.5, random_state = 10101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
print(f"Tip = {model.intercept_[0]:.3f} + " +
      f"{model.coef_[0][1]:.3f} Lunch + " +
      f"{model.coef_[0][2]:.3f} Total Bill + " +
      f"{model.coef_[0][3]:.3f} Size")
```

```python
results = model.predict(x_test)
results.shape
```

```python
y_test.shape
```

```python
score = model.score(x_test, y_test)
print (f"LR model score = {score: .3f}")
```
